```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
ticker="META"
df=yf.download(ticker,start="2020-01-01",end="2024-01-01")
print(df.head())
plt.plot(df['Close'])
plt.xlabel("Date")
plt.ylabel("Share price")
plt.title("META Share price fluctuations",color="Orange")
```
