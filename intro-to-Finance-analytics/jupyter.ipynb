ASSIGNMENT-2: walmart Data analysis

```python
import pandas as pd
df=pd.read_csv('/Walmart.csv')
print(df.head())
```

```python
print(df.dtypes)
```

```python
print("Mean of weekly sales is :",df['Weekly_Sales'].mean())
print("Mean of temperatures is :",df['Temperature'].mean())
print("Mean of fuel price is :",df['Fuel_Price'].mean())
print("Mean of CPI is :",df['CPI'].mean())
print("Mean of Unemployment is :",df['Unemployment'].mean())
```

```python
print("Median of weekly sales is :",df['Weekly_Sales'].median())
print("Median of temperatures is :",df['Temperature'].median())
print("Median of fuel price is :",df['Fuel_Price'].median())
print("Median of CPI is :",df['CPI'].median())
print("Median of Unemployment is :",df['Unemployment'].median())
```

```python
print("Mode of weekly sales is :",df['Weekly_Sales'].mode())
print("Mode of temperatures is :",df['Temperature'].mode())
print("Mode of fuel price is :",df['Fuel_Price'].mode())
print("Mode of CPI is :",df['CPI'].mode())
print("Mode of Unemployment is :",df['Unemployment'].mode())
```

```python

print("Variance of weekly sales data is :",df['Weekly_Sales'].var())
print("Standard deviation of weekly sales data is :",df['Weekly_Sales'].std())
print("Variance of Temperature data is :",df['Temperature'].var())
print("Standard deviation of Temperature data is :",df['Temperature'].std())
print("Variance of Fuel price data is :",df['Fuel_Price'].var())
print("Standard deviation of Fuel price data is :",df['Fuel_Price'].std())
print("Variance of CPI data is :",df['CPI'].var())
print("Standard deviation of CPI data is :",df['CPI'].std())
print("Variance of unemployment data is :",df['Unemployment'].var())
print("Standard deviation of Unemployment data is :",df['Unemployment'].std())
```

```python
plt.hist(df['Weekly_Sales'],color="Orange")
plt.show()
```

```python
plt.boxplot(df['Weekly_Sales'])
plt.title("Box plot for weekly sales of Walmart")
plt.show()
```

```python
print(df['Weekly_Sales'].min())
q1_value=df['Weekly_Sales'].quantile(0.25)
q2_value=df['Weekly_Sales'].quantile(0.50)
print(df['Weekly_Sales'].median())
print(q1_value)
print(q2_value)
```
